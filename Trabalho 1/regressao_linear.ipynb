```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Importando o dataset

with open("datasetFase1.txt", "r") as f:
    data = f.read()

data = data.replace("[", "").replace("]", "").replace(";", ",")
variables = {}

exec(data, variables)

x = np.array(variables["x1"])
y = np.array(variables["y1"])
```

```python
# Implementação da função correlacao

def correlacao(x, y):

    n = len(x) # retorna lista
    media_x = sum(x) / n # n total de elementos
    media_y = sum(y) / n

    numerador = sum((x[i] - media_x) * (y[i] - media_y) for i in range(n))
    denominador_x = sum((x[i] - media_x) ** 2 for i in range(n))
    denominador_y = sum((y[i] - media_y) ** 2 for i in range(n))

    return numerador / (denominador_x * denominador_y) ** 0.5
```

```python
# Implementação da função regressao

def regressao(x, y):
    n = len(x) # n elementos
    media_x = sum(x) / n # media
    media_y = sum(y) / n

    numerador = sum((x[i] - media_x) * (y[i] - media_y) for i in range(n))
    denominador = sum((x[i] - media_x) ** 2 for i in range(n))

    beta1 = numerador / denominador
    beta0 = media_y - beta1 * media_x

    return beta0, beta1
```

```python
# Implementação da função demo

def demo(x, y):
    r = correlacao(x, y)
    beta0, beta1 = regressao(x, y)

    plt.scatter(x, y, color='blue', label='Dados')

    # Criar um intervalo para suavizar a linha de regressão
    x_range = np.linspace(min(x), max(x), 100)
    y_pred = beta0 + beta1 * x_range
    plt.plot(x_range, y_pred, color='red', label='Regressão Linear')

    plt.title(f"Correlação: {r:.4f}, Regressão: β0={beta0:.4f}, β1={beta1:.4f}")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.legend()
    plt.show()
```

```python
# Executando a função demo para o dataset 1

# x1 = [10;8;13;9;11;14;6;4;12;7;5];
# y1 = [8.04;6.95;7.58;8.81;8.33;9.96;7.24;4.26;10.84;4.82;5.68];

demo(np.array(variables["x1"]), np.array(variables["y1"]))
plt.show()
```

```python
# Executando a função demo para o dataset 2

# x2 = [10;8;13;9;11;14;6;4;12;7;5];
# y2 = [9.14;8.14;8.47;8.77;9.26;8.10;6.13;3.10;9.13;7.26;4.74];

demo(np.array(variables["x2"]), np.array(variables["y2"]))
plt.show()
```

```python
# Executando a função demo para o dataset 3

# x3 = [8;8;8;8;8;8;8;8;8;8;19];
# y3 = [6.58;5.76;7.71;8.84;8.47;7.04;5.25;5.56;7.91;6.89;12.50];

demo(np.array(variables["x3"]), np.array(variables["y3"]))
plt.show()
```
